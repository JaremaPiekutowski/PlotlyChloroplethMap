# Plotly Chloropleth Map

## IMPORTS

```python
import json
import pandas as pd
import plotly.express as px
import plotly.io as pio
```

## GET MAP

```python
wojewodztwa = json.load(open("wojewodztwa-medium.geojson"))
```

## GET DATA

```python
df = pd.read_excel('prosiaki.xlsx')
```

```python
df.head()
```

```python
fig = px.choropleth(df, locations='id', geojson=wojewodztwa, color="data", scope='europe')
fig.update_geos(fitbounds="locations", visible=False);
```

```python
fig.show()
```
